```haskell
:extension NoMonomorphismRestriction FlexibleContexts TypeFamilies
import Control.Monad
import Control.Monad.State

import System.Random
import Data.Random
import Data.Random.Distribution.Pareto
import Data.RVar

import Diagrams.Prelude hiding (normal)
import Data.Colour.Palette.BrewerSet
import qualified Diagrams.Color.XKCD as XKCD
import Data.Colour (withOpacity)

import Data.Tuple
```

```haskell
sampleAndDraw seed randomDiagram = withImgHeight 500 $ diagram d
  where d = evalState (sampleRVar randomDiagram) $ mkStdGen seed
```

```haskell
scanM :: (Monad m) => (a -> b -> m a) -> a -> [b] -> m [a]
scanM f q [] = return [q]
scanM f q (x:xs) =
   do q2 <- f q x
      qs <- scanM f q2 xs
      return (q:qs)

iterateM n f a = scanM (\x _ -> f x) a $ replicate n ()

-- example usage
incrementNegsOnly x = if x < 0 then Just (x + 1) else Nothing

iterateM 4 incrementNegsOnly (-4)
iterateM 4 incrementNegsOnly (-3)
```

```haskell
randomLine width var = replicateM width $ normal 0.005 var

drawLine line balance = (cubicSpline False (map p2 $ zip [0,1..] line))
  # lcA (blend balance blue yellow `withOpacity` 0.2)
  # lw 3
                                                             
walkingLine n width var = do
  steps <- replicateM n $ randomLine width var
  let colors = map (\x -> fromIntegral x / fromIntegral n) [1,2..n]
  let walk = scanl1 (zipWith (+)) steps
  return $ mconcat $ zipWith drawLine walk colors
```

```haskell
sampleAndDraw 2 $ walkingLine 1000 15 0.05
```

```haskell
randomVec var = curry r2 <$> (normal 0 var) <*> (normal 0 var)

-- Shape -> Rand Shape
perturb var shape = do
  deltas <- replicateM (length shape) $ randomVec var
  return $ zipWith (.+^) shape deltas
  
evolve steps var shape = iterateM steps (perturb var) shape
```

```haskell
drawCurve points = (cubicSpline True points)
  # lcA (blue `withOpacity` 0.2)
  # lw 3

morphingHexagon = do
  evolution <- evolve 200 0.02 $ square 1
  return $ mconcat $ map drawCurve evolution
  
sampleAndDraw 2 morphingHexagon
```
